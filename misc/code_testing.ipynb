```julia
rosenbrock(x, y) = 100*(y - x^2)^2 + (x - 1)^2
wood(x) = 100*(x[2] - x[1]^2)^2 + (1 - x[1]) + 90*(x[4] - x[3]^2)^2 + (1 - x[3])^2 + 10*(x[2] + x[4] - 2)^2 + 0.1*(x[2] - x[4])^2
freudesteinRoth(x) = (-13 + x[1] + ((5 - x[2]) * x[2] - 2) * x[2])^2 + (-29 + x[1] + ((x[2] + 1) * x[2] - 14) * x[2])^2
freudesteinRoth(x, y) = (-13 + x + ((5 - y) * y - 2) * y)^2 + (-29 + x + ((y + 1) * y - 14) * y)^2
booth(x, y) = (x + 2*y - 7)^2 + (2*x + y - 5)^2
mccormick(x, y) = sin(x + y) + (x - y)^2 - (3/2)*x + (5/2) + 1
```

```julia
x = -100:0.5:100
y = -100:0.5:100

p1 = contour(x, y, rosenbrock, fill = true)
p2 = contour(x, y, booth, fill = true)
p3 = contour(x, y, mccormick, fill = true)
p4 = contour(x, y, freudesteinRoth, fill = true)

plot(p1)
```
